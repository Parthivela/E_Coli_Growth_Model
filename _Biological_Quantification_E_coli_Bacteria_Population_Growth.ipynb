```python
#import the required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#take the x_values which is nothing but time(hours)
growth_hours=[1,2,6,7,8,9,11,12] # take it as a list here we can give the range also but i have used the specific values given in the module_1
#calculate the bacterial population(growth) by using the given formula i.e y=2**x
bacteria_population=[2**x for x in growth_hours]
#print the calculated bacterial population for different hours
print("Hours   ---  E-coli Bacteria Population")
#print the values
for hour,population in zip(growth_hours,bacteria_population):
    print(f"{hour}       ---      {population}")
#calcuate the bins for Histogram by using Sturge's Rule:
histogram_bin_count=int(1+ np.log2(len(bacteria_population)))
#print the bincount
print("Histogram Bin Count:",histogram_bin_count)
#plot the Histogram with relative frequency using seaborn
plt.figure(figsize=(7,5))
sns.histplot(bacteria_population,bins=histogram_bin_count,kde=True,color="blue",edgecolor="yellow",linewidth=1.5,alpha=0.7)
#give the labels
plt.xlabel("No of E-coli Bacteria Cells",fontsize=12,fontweight='bold',color='darkblue')
plt.ylabel("Relative Frequency",fontsize=12,fontweight='bold',color='darkred')
plt.title("Histogram of E-coli Bacteria Population Growth",fontsize=14,fontweight='bold',color='darkgreen')
plt.grid(axis='y',linestyle='dashdot',color='gray')
plt.show()

```
